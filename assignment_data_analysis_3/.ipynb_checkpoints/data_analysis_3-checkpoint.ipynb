```python
### Definition of the data we need for the PCA
genes = df.columns.values
X = df.iloc[:, 0:47]
```

```python
### Initiation of PCA and applying it to our data
pca = sklearn.decomposition.PCA()
pca.fit(X)
```

```python
### Getting the explained_variance ratio
pca.explained_variance_ratio_
```

```python
## Plot the explained_variance_ratio_
sns.barplot(x=np.arange(0, 47), y=pca.explained_variance_ratio_)
plt.title("PC Explained Variance - pd Dataset")
plt.xlabel("principal components")
plt.ylabel("%of explained variance")
```

```python
### Applying the dimensionality reduction to our data 
reduce_df = pca.transform(X)
```

```python
### Making a dataframe and add the label column
reduce_df = pd.DataFrame(reduce_df)
```

```python
reduce_df
```

```python
### Ploting the results of the reduction
sns.scatterplot(data= reduce_df, x=0, y=1) 
sns.despine()
plt.title ("PCA reduction")
plt.xlabel("PCA 0")
plt.ylabel("PCA 1")
```

```python
### Add the cluster column
reduce_df ["cluster"] = kmeans.labels_
reduce_df
```

```python
### Ploting the cluster and the labels
sns.relplot (data=reduce_df, x=0, y=1, col="cluster")
sns.despine()
plt.title ("PCA reduction")
plt.xlabel("PCA0")
plt.ylabel("PCA1")
```
